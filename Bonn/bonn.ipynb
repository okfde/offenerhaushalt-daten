```python
df = pd.read_csv('C:\\Users\\Michael\\Documents\\OKFN\\Offener Haushalt\\Bonn\\Sammeldatei_2017.csv',  sep=';', encoding='cp437') 
```

```python
df.Prctr = df.Prctr.astype(str)
```

```python
df.Konto = df.Konto.astype(str)
```

```python
# Festlegen welche Haushaltstitel Einnahmen (1) oder Ausgaben (2) sind 
dt = {
1 : 'Innere Verwaltung',
2 : 'Sicherheit und Ordnung',
3 : 'Schulträgeraufgaben',
4 : 'Kultur und Wissenschaft',
5 : 'Soziale Leistungen',
6 : 'Kinder-, Jugend- und Familienhilfe',
7 : 'Gesundheitsdienste',
8 : 'Sportförderung',
9 : 'Räumliche Planung und Entwicklung',
10 : 'Bauen und Wohnen',
11 : 'Ver- und Entsorgung',
12 : 'Verkehrsflächen, -anlagen, ÖPNV',
13 : 'Natur und Landschaftspflege',
14 : 'Umweltschutz',
15 : 'Wirtschaft und Tourismus',
16 : 'Allgemeine Finanzwirtschaft',
17 : 'Stiftungen',
0 : 'Produktbereichsübergreifend' 
}
```

```python
df['Produktbereich'] = [dt[int(i[5:7])] for i in df['Prctr']]
```

```python
with open ("C:\\Users\\Michael\\Documents\\OKFN\\Offener Haushalt\\Bonn\\produktgruppecode.csv", encoding = 'utf-8') as infile: 
    reader = csv.reader(infile)
    produktgruppecode = {line[0]:line[1] for line in reader}
```

```python
df['Produktgruppe'] = [produktgruppecode[i] for i in df['Prctr']]
```

```python
with open ("C:\\Users\\Michael\\Documents\\OKFN\\Offener Haushalt\\Bonn\\kostenartencode.csv", encoding = 'utf-8') as infile: 
    reader = csv.reader(infile)
    kostenartencode = {line[0]:line[1] for line in reader}
```

```python
df['Kostenart'] = [kostenartencode[i] for i in df['Konto']]
```

```python
df.head()
```

```python
df['Betrag'] = df['      in PrCtrHW']
```

```python
del df['      in PrCtrHW']
```

```python
df['Betrag'] = df['Betrag'].str.replace(',','.')
```

```python
df.Betrag = df.Betrag.astype(float)
```

```python
df = df[(df['Konto'].str[0] == '4') | (df['Konto'].str[0] == '5')]
```

```python
df['Budget_Richtung'] = ['Erträge' if i[0] == '4' else 'Aufwendungen' for i in df['Konto']]
```

```python
df[df['Budget_Richtung']== 'Erträge']
```

```python
df['Betrag'] = df.apply(lambda row: -1*(row["Betrag"]) if row["Budget_Richtung"] == "Erträge" else row["Betrag"], axis=1)
```

```python
df['Plan_Ist'] = 'Plan'
```
